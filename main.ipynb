```python
import pandas as pd
import numpy as np
import csv
import process
import tui

#initializing function
def read_data(file_path):
    tui.started(f"Reading data from {file_path}")
    #reading data into a dataframe and assigning header as 0th row
    df = pd.read_csv(file_path, header=0)
    tui.completed()
    return df

def run():
    cd = read_data("CarPrice.csv")
    print(type(cd))  # Output: <class 'pandas.core.frame.DataFrame'>
    # initiating an infinite white statement to prompt the user for selection
    while True:
        selection = tui.menu()
        if selection == "car":
            process.car(cd)
        elif selection == "cylinder_num":
            process.cylinder_num(cd)
        elif selection == "carbody":
            process.carbody(cd)
        elif selection == "columns_specified_2_5":
            process.columns_specified_2_5(cd)
        elif selection == "carnames_alphabetically":
            process.carnames_alphabetically(cd)
        elif selection == "car_body_prices":
            process.car_body_prices(cd)
        elif selection == "top_5_car_sale":
            process.top_5_car_sale(cd)
        elif selection == "self_selected_data":
            process.self_selected_data(cd)
        elif selection == "cars_per_fuelsystem":
            process.cars_per_fuelsystem(cd)
        elif selection == "hp_5cheapest_cars":
            process.hp_5cheapest_cars(cd)
        elif selection == "buying_patterns":
            process.buying_patterns(cd)
        elif selection == "exit":
            break
        else:
            tui.error("Invalid Selection!")
# initiating conditional statement to run the program
if __name__ == "__main__":
    run()                                                                                                                 
```
